## Formula para calcular el area
$$A = \frac{b \cdot h}{2}$$

# Datos de entrada
* b = ?
* h = ?
---
## Diagrama de flujo de LucidChart

```python
# Calcaular el area de un triangulo
# Variables: b (base), h (altura), A (area)
# Formula: A = (base * altura) / 2
# Capturar los valores de b y h

print("Calcular el area de un triangulo")

b = int(input("Ingrese el valor de la base: "))
h = int(input("Ingrese el valor de la altura: "))

# Aplicamos la formula A = (b * h) / 2
A = (b * h) / 2

print("El area de un triangulo es: ", A)
print("Programa realizado por Jonathan Baray NL = 06")
```
